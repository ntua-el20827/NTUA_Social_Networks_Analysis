```python
%matplotlib inline
import matplotlib
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random
```

### Α) Κατασκευή και οπτικοποίηση σύνθετων τύπων δικτύου

### Τοπολογία REG για βαθμό πλέγματος $d = 4$

```python
random.seed(seedno)
#Size of Network
n=47
d=4
# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
REG = nx.watts_strogatz_graph(n,d,0,seed=seedno)
#n -> The number of nodes
#d -> Each node is joined with its d nearest neighbors in a ring topology.
#p -> The probability of rewiring each edge
# If p = 0, the graph is a ring lattice.
# If p = 1, all edges are rewired, resulting in a random graph.
#seed -> random_state, or None (default) Indicator of random number generation state.

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG, with_labels=True)
```

Παρότι δεν είναι εμφανές στο πάνω σχήμα, κάθε κόμβο συνδέεται με τους 2 επόμενους και τους 2 προηγούμενους. Για παραδειγμα ο κόμβος 0 συνδέεται με τους κόμβους 1, 2, 46 και 45. Για να γίνει πιο εμφανές πρέπει να δείξουμε ενα γράφημα με μικρότερο πλήθος κόμβων.
Επιλέγουμε p=0 γιατί θέλουμε να δημιουργήσουμε ένα πλήρες πλέγμα (ring lattice) με βαθμό 4.

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
random.seed(seedno)
n=47
M = 168
### Αλλάζω τον αριθμό των ακμών για να προκύπτει συνδεδμένος γράφος.
### Αν το αφήναμε 150, επειδή n=47, η συνάρτηση nx.is_connected(RGER) θα επέστρεφε False!!
RGER = nx.gnm_random_graph(n,M, seedno)
#n -> The number of nodes
#M -> The number of edges
#seed -> random_state, or None (default) Indicator of random number generation state.


# οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER, with_labels=True)
```

Σε αυτό το δίκτυο ορίζω το πληθος των κόμβων και των ακμών και η δημιοιυργία του γίνεται με τυχαίο τρόπο.

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
random.seed(seedno)
n=47
R=0.35

# για την κατασκευή του RGG χρησιμοποιήστε την random_geometric_graph από το networkx
RGG= nx.random_geometric_graph(n,R,seed=seedno)
#n -> The number of nodes
#R -> nodes are uniformly placed in the unit cube and two nodes are connected if the Euclidean distance between the nodes is at most radius.

#οπτικοποίηση: https://networkx.org/documentation/stable/auto_examples/drawing/plot_random_geometric_graph.html
# position is stored as node attribute data for random_geometric_graph
pos = nx.get_node_attributes(RGG, "pos")

# find node near center (0.5,0.5)
dmin = 1
ncenter = 0
for n in pos:
    x, y = pos[n]
    d = (x - 0.5) ** 2 + (y - 0.5) ** 2
    if d < dmin:
        ncenter = n
        dmin = d

# color by path length from node near center
p = dict(nx.single_source_shortest_path_length(RGG, ncenter))

plt.figure(figsize=(8, 8))
nx.draw_networkx_edges(RGG, pos, alpha=0.4)
nx.draw_networkx_nodes(
    RGG,
    pos,
    nodelist=list(p.keys()),
    node_size=80,
    node_color=list(p.values()),
    cmap=plt.cm.Reds_r,
)

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)
plt.axis("off")
plt.show()

# Για την οπτικοποίηση του RGG χρησιμοποιήσα τον κώδικα απο το link που δώθηκε. 
```

Εδω  δημιουργώ ένα τυχαίο γεωμετρικό γράφημα με κόμβους που τοποθετούνται στο επίπεδο και συνδέονται μεταξύ τους αν η απόσταση τους είναι μικρότερη από την ακτίνα σύνδεσης.

### Τοπολογία SF με παράμετρο επισύναψης $d=4$

```python
random.seed(seedno)
n=47 
d=4

# για την κατασκευή του SF χρησιμοποιήστε την barabasi_albert_graph από το networkx
SF = nx.barabasi_albert_graph(n,d,seed=seedno) 
#n -> The number of nodes
#d -> Number of edges to attach from a new node to existing nodes
#seed -> random_state, or None (default) Indicator of random number generation state.

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SF, with_labels=True)
```

Στο Scale-free δίκτυο αυτό ορίζω το πλήθος των κόμβων και τον αριθμό των συνδέσεων που θα έχει ο κάθε νέος κόμβος που προστίθεται στο δίκτυο. Ώστόσο κάθε νεός κόμβος που προστίθεται στο δίκτυο συνδέεται με τους κόμβους που έχουν τους μεγαλύτερους βαθμούς.

Ετσι δημιουργείται ένα δίκτυο που ακολουθεί την κατανομή της εξουσίας (power law), δηλαδή λίγοι κόμβοι έχουν υψηλό βαθμό διασύνδεσης (hubs), ενώ οι περισσότεροι κόμβοι έχουν μικρότερο βαθμό. Αυτό το δίκτυο αναπαριστά τις ιδιότητες των πραγματικών κοινωνικών και τεχνολογικών δικτύων, όπου λίγοι κόμβοι (π.χ., influencers σε social media) έχουν πολύ περισσότερες συνδέσεις από τους υπόλοιπους.

### Τοπολογία SW με βαθμό αρχικού πλέγματος $d=4$ και πιθανότητα ανασύνδεσης $p=0.25$

```python
random.seed(seedno)
#n=4x
d=4
p=0.25

# για την κατασκευή του SW χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
SW = nx.watts_strogatz_graph(n,d,p,seed=seedno)
#n -> The number of nodes
# d -> Each node is joined with its d nearest neighbors in a ring topology.
# p -> The probability of rewiring each edge
# If p = 0, the graph is a ring lattice.
# If p = 1, all edges are rewired, resulting in a random graph.
# seed -> random_state, or None (default) Indicator of random number generation state. 

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SW, with_labels=True)
```

Στο Small-world (SW) δίκτυο αυτο ορίζω το πλήθος των κόμβων και τον βαθμό του αρχικού πλέγματος. Στη συνέχεια, κάθε ακμή του αρχικού πλέγματος ανασυνδέεται με πιθανότητα p. Η ανασύνδεση αυτή δημιουργεί τοπικές δομές σε μια τυχαία τοπολογία, ενώ διατηρεί την τοπική συνδεσιμότητα του αρχικού πλέγματος. 

Αυτό το δίκτυο αναπαριστά τις ιδιότητες των πραγματικών δικτύων, όπου οι κόμβοι συνδέονται τόσο τοπικά όσο και τυχαία.

Επομένως, συγκεντρωτικά, θα ασχοληθούμε με τα ακόλουθα διαγράμματα τοπολογιών δικτύων:

REG με d=4: Το τακτικό αυτό γράφημα (Regular Graph) έχει κάθε κόμβο να συνδέεται με τους δύο προηγούμενους και τους δύο επόμενους κόμβους του. Αυτή η τοπολογία είναι κατάλληλη για την προσομοίωση των δικτύων όπου η τοπική συνδεσιμότητα είναι σημαντική, όπως στα τηλεπικοινωνιακά δίκτυα.

RG (ER): Σε αυτό το τυχαίο γράφημα (Random Graph, τύπου Erdős-Rényi), ορίζεται ο συνολικός αριθμός κόμβων και ακμών και οι συνδέσεις σχηματίζονται τυχαία. Αυτό το μοντέλο δεν έχει συγκεκριμένη δομή, καθώς οι ακμές τοποθετούνται τυχαία ανάμεσα στους κόμβους, οδηγώντας σε μια ομοιόμορφη πιθανότητα σύνδεσης μεταξύ τους.

RGG με R=0.35: Το Τυχαίο Γεωμετρικό Γράφημα (Random Geometric Graph) δημιουργείται με κόμβους που τοποθετούνται τυχαία στο επίπεδο, ενώ οι σύνδεσεις σχηματίζονται μόνο αν η απόσταση μεταξύ δύο κόμβων είναι μικρότερη από την καθορισμένη ακτίνα R=0.35. Αυτή η τοπολογία είναι κατάλληλη για την προσομοίωση ασύρματων και φυσικών δικτύων όπου η εγγύτητα είναι καθοριστικός παράγοντας για τις συνδέσεις.

SF με d=4: Το Scale-Free δίκτυο βασίζεται στην ιδέα της προτιμησιακής σύνδεσης, όπου κάθε νέος κόμβος που προστίθεται συνδέεται με προτίμηση προς κόμβους με υψηλό βαθμό. Στο συγκεκριμένο δίκτυο, κάθε νέος κόμβος προσθέτει τέσσερις συνδέσεις με ήδη υπάρχοντες κόμβους υψηλού βαθμού, δημιουργώντας λίγους κόμβους με πολύ υψηλό βαθμό διασύνδεσης (hubs) και πολλούς με χαμηλότερο. Αυτή η δομή ακολουθεί την κατανομή της εξουσίας (power law) και απαντά σε κοινωνικά και τεχνολογικά δίκτυα.

SW με d=4 και p=0.25: Το Small-World δίκτυο αρχικά διαμορφώνεται ως τακτικό πλέγμα, με κάθε κόμβο να συνδέεται με τέσσερις γειτονικούς του (d=4). Στη συνέχεια, κάθε ακμή ανασυνδέεται με πιθανότητα p=0.25, δημιουργώντας μακρινές συνδέσεις που αυξάνουν την τυχαιότητα, χωρίς να χάνεται η τοπική συνδεσιμότητα. Αυτή η τοπολογία μιμείται την ταυτόχρονη ύπαρξη τοπικών και τυχαίων συνδέσεων, χαρακτηριστικό που συναντάται σε πραγματικά κοινωνικά δίκτυα και επιτρέπει την ταχύτερη διάδοση πληροφορίας.

### B) Μετρικές δικτύου

### B1. Συντελεστής ομαδοποίησης (Clustering coefficient)

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β1.1 Να υπολογιστεί ο συντελεστής ομαδοποίησης κάθε κόμβου και να αναπαρασταθεί η κατανομή του.
#### Β1.2  Να υπολογιστεί ο μέσος συντελεστής ομαδοποίησης και η διασπορά του.
#### Β.1.3 Να βρεθούν οι κόμβοι με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης. 
#### Να σχολιαστούν τα αποτελέσματα.

Λιγα Σχολια για τον Συντελεστή Ομαδοποίησης:


Ουσιαστικά αυτος ο συντελεστής δείχνει το πόσο πιθανό είναι οι γείτονες ενός κόμβου να συνδέονται και μεταξύ τους. Αντιπροσωπεύει, δηλαδή, την "κλειστότητα" των συνδέσεων γύρω από κάθε κόμβο, δίνοντας μια ένδειξη της ύπαρξης υπο-κοινοτήτων ή ομάδων στο δίκτυο. 


Παίρνει τιμές μεταξύ 0 και 1. 
0 -> κανένας από τους γείτονες του κόμβου δεν είναι συνδεδεμένοι μεταξύ τους.
1 -> όλοι οι γείτονες του κόμβου είναι συνδεδεμένοι μεταξύ τους.

```python
# Usefull functions
def cc(G):
    return nx.clustering(G)
def av_cc(G):
    return nx.average_clustering(G)

#Graphs
graphs = [REG, RGER, RGG, SF, SW]
graph_names = ['REG', 'RGER', 'RGG', 'SF', 'SW']
```

```python
#Β1.1
for graph, graph_name in zip(graphs, graph_names):
    #calculate the clustering coefficient for each node
    clustering_coeffs = cc(graph) #this is a dictionary! (node: clustering coefficient)
    coeff_values = list(clustering_coeffs.values())
    
    # Plot distribution of clustering coefficients
    plt.hist(coeff_values, bins=20, alpha=0.75, range=(0, 1))
    plt.title(f'Clustering Coefficient Distribution for {graph_name}')
    plt.xlabel('Clustering Coefficient')
    plt.ylabel('Frequency')
    plt.show()
```

Σχολια για το Β1.1:

REG: Λόγω της ομοιόμορφης δομής της τοπολογίας, όλοι οι κόμβοι έχουν τον ίδιο συντελεστή ομαδοποίησης. Αυτός όμως δεν ειναι ίσος με τον βαθμό του κάθε κόμβου! (d=4 ενώ cc=0.5) 

RG(ER): Ο συντελεστής ομαδοποίησης είναι χαμηλός, το οποίο είναι λογίκό αν λαβομε υπόψην την τυχαιότητα των ακμών. Αυτό σημαίνει ότι οι γείτονες ενός κόμβου δεν είναι πιθανό να συνδέονται μεταξύ τους.

RGG: Ο συντελεστής ομαδοποίησης είναι υψηλός, καθώς οι κόμβοι συνδέονται μόνο αν η απόσταση τους είναι μικρότερη από την ακτίνα σύνδεσης. Αυτό σημαίνει ότι οι γείτονες ενός κόμβου είναι πιθανό να συνδέονται μεταξύ τους.

SF(BA): Σε αυτά τα δίκτυα, οι κόμβοι υψηλού βαθμού (hubs) συνδέονται κυρίως με κόμβους χαμηλότερου βαθμού και όχι με άλλους hubs. Αυτό σημαίνει ότι οι hubs συνήθως δεν σχηματίζουν τριγωνικές συνδέσεις, με αποτέλεσμα οι περισσότεροι κόμβοι να έχουν χαμηλό συντελεστή ομαδοποίησης, καθώς οι νέοι κόμβοι συνδέονται με τους κόμβους υψηλού βαθμού. Αυτό σημαίνει ότι οι γείτονες ενός κόμβου δεν είναι πιθανό να συνδέονται μεταξύ τους.

SW: Αυτη η τοπολογία είναι ιδιαίτερα ενδιαφέρουσα καθώς συνδυάζει τα χαρακτηριστικά της τοπικής συσσωμάτωσης με εκείνα της τυχαιότητας. Ετσι, το SW μοντέλο καταφέρνει έτσι να διατηρήσει τόσο ισχυρά τοπικά clusters όσο και μεγάλες αποστάσεις συνδέσεων, επιτυγχάνοντας τη διάδοση πληροφορίας με αποδοτικό τρόπο.

```python
#Β.1.2
for graph, graph_name in zip(graphs, graph_names):
    avg_clustering = av_cc(graph)
    clustering_coeffs = list(nx.clustering(graph).values())
    variance_clustering = np.var(clustering_coeffs)
    
    print(f'{graph_name} - Average Clustering Coefficient: {avg_clustering:.4f}')
    print(f'{graph_name} - Clustering Coefficient Variance: {variance_clustering:.4f}\n')
```

Σχολια για το Β1.2:

REG: Το αποτέλεσμα αυτό είναι αναμενόμενο για ένα REG γράφημα, καθώς όλοι οι κόμβοι έχουν τον ίδιο βαθμό διασύνδεσης και συμμετρικές συνδέσεις με τους γείτονές τους.

RG(ER): Η χαμηλή μέση τιμή του συντελεστή ομαδοποίησης είναι χαρακτηριστική για τα τυχαία γραφήματα, όπου η πιθανότητα τριγωνικών συνδέσεων είναι μικρή. Η μικρή διασπορά (0.0123) δείχνει ότι το clustering είναι λίγο πολύ ομοιόμορφο, καθώς η τυχαιότητα στις συνδέσεις δεν ενισχύει ιδιαίτερα τη δημιουργία τριγώνων.

RGG: Το RGG εμφανίζει υψηλό μέσο συντελεστή ομαδοποίησης, κάτι που είναι πολυ λογικό, καθώς οι κόμβοι συνδέονται όταν βρίσκονται κοντά στον χώρο. Η αρκετά χαμηλή διασπορά (0.0003) δείχνει ότι οι τιμές του clustering είναι συγκεντρωμένες γύρω από τη μέση τιμή.

SF(BA): Ο μέσος συντελεστής ομαδοποίησης είναι χαμηλός, καθώς οι hubs συνδέονται κυρίως με κόμβους χαμηλότερου βαθμού και όχι με άλλους hubs. Η χαμηλή διασπορά (0.0001) δείχνει ότι οι τιμές του clustering είναι συγκεντρωμένες γύρω από τη μέση τιμή.

SW: Στο Small-World δίκτυο, το μέσο clustering είναι μέτριο και σχετικά χαμηλό σε σχέση με το RGG, λόγω της ανασύνδεσης των ακμών που δημιουργεί απομακρυσμένες συνδέσεις. Ωστόσο, είναι υψηλότερο από το RG λόγω της τοπικής συνδεσιμότητας που διατηρείται στο αρχικό πλέγμα. Η σχετικά υψηλή διασπορά (σε σύγκριση με τις προηγούμενες) δείχνει ότι υπάρχει μια μεγαλύτερη ποικιλία στο clustering ανάμεσα στους κόμβους. Αυτό απορρέει από τη σύνθεση τοπικών clusters με πιο τυχαίες συνδέσεις, κάτι που ενισχύει την ποικιλομορφία στη συσσωμάτωση


```python
#Β.1.3
for graph, graph_name in zip(graphs, graph_names):
    clustering_coeffs = cc(graph)
    min_cc = min(clustering_coeffs.values())
    max_cc = max(clustering_coeffs.values())
    
    min_cc_nodes = [node for node, coeff in clustering_coeffs.items() if coeff == min_cc]
    max_cc_nodes = [node for node, coeff in clustering_coeffs.items() if coeff == max_cc]
    
    print(f'{graph_name} - Nodes with smallest clustering coefficient ({min_cc}): {min_cc_nodes}')
    print(f'{graph_name} - Nodes with largest clustering coefficient ({max_cc}): {max_cc_nodes}\n')
```

### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

RG(ER): Το μέσο μήκος μονοπατιού είναι χαμηλό, καθώς οι κόμβοι συνδέονται τυχαία και η πιθανότητα να υπάρχει μονοπάτι μεταξύ δύο κόμβων είναι υψηλή. Αυτό σημαίνει ότι το δίκτυο είναι συνδεδεμένο και οι αποστάσεις μεταξύ των κόμβων είναι σύντομες.

RGG: Το μέσο μήκος μονοπατιού είναι μέτριο, καθώς οι κόμβοι συνδέονται μόνο αν βρίσκονται κοντά στον χώρο. Αυτό σημαίνει ότι το δίκτυο είναι συνδεδεμένο, αλλά οι αποστάσεις μεταξύ των κόμβων είναι μεγαλύτερες από το RG. Εδω πρέπει να δώσουμε περισσότερη εμφαση στην υψηλή τιμή του Variance, το οποίο υποδηλώνει ισχυρές τοπικές ομάδες, οπως ακριβώς είδαμε και στο clustering coefficient.

SF(BA): Το μέσο μήκος μονοπατιού είναι το χαμηλότερο που παρατηρούμε, καθώς οι hubs συνδέονται με κόμβους χαμηλότερου βαθμού και όχι με άλλους hubs.Επομένως οι κεντρικοί κόμβοι (hubs) επιτρέπουν γρήγορες συνδέσεις σε όλο το δίκτυο.

SW: Εδώ, το μήκος είναι μέτριο, αλλά καλύτερο από το REG, υποδεικνύοντας ότι οι συνδέσεις "μικρού κόσμου" μειώνουν τα μονοπάτια μεταξύ απομακρυσμένων κόμβων.

Σχόλια για το Β2.2:
- Diameter: Η διάμετρος είναι η μέγιστη απόσταση μεταξύ δύο κόμβων στο δίκτυο. Αντιπροσωπεύει τη μέγιστη απόσταση που πρέπει να διανύσει ένα μήνυμα για να φτάσει από τον ένα κόμβο στον άλλο. Η διάμετρος είναι ένας σημαντικός δείκτης της συνδεσιμότητας του δικτύου.
- Radius:  Η ακτίνα ενός δικτύου είναι η ελάχιστη απόσταση (σε αριθμό ακμών) από τον πιο απομακρυσμένο κόμβο προς τον πιο κοντινό κεντρικό κόμβο (ή "κεντρικό σημείο").Αναδεικνύει την ελάχιστη απόσταση που χρειάζεται για να φτάσει κανείς σε όλους τους άλλους κόμβους του δικτύου από την "κεντρική" τοποθεσία. Όσο μικρότερη είναι η ακτίνα, τόσο πιο προσβάσιμο είναι το δίκτυο.

Με βάση τα παραπάνω, παρατηρούμε τα εξής:

- Το SW έχει μικρότερη διάμετρο από το REG, καθώς οι συνδέσεις "μικρού κόσμου" μειώνουν τα μονοπάτια μεταξύ απομακρυσμένων κόμβων.
- Το RGG έχει τη μεγαλύτερη διάμετρο, καθώς οι κόμβοι συνδέονται μόνο αν βρίσκονται κοντά στον χώρο.


#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

```python
def find_nodes(G):
    diameter = nx.diameter(G)
    path_lengths = compute_path_lengths(G)

    #Dictionaries to store the number of connections for each node
    shortest_paths = {}

    for node in range(len(path_lengths)):
        shortest_paths[node] = sum(1 for length in path_lengths[node].values() if length == diameter)
    
    #Find the node with the most connections in the diameter
    max_connections = max(shortest_paths.values())
    nodes = [node for node, connections in shortest_paths.items() if connections == max_connections]
    return nodes, max_connections, diameter
    

for graph, graph_name in zip(graphs, graph_names):
    nodes,max_connections,diameter = find_nodes(graph)
    print(f'{graph_name} - Nodes with most connections in the diameter: {nodes} with {max_connections} connections of length {diameter}\n ')
```

#### Περιφέρεια του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με τη διάμετρό του. 
#### Κέντρο του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με την ακτίνα του.  
#### Β.2.4 Για κάθε τοπολογία, να βρεθεί η περιφέρεια και το κέντρο της και να σχολιαστούν τα αποτελέσματα.

```python
def find_periphery(G):
    #using eccentricity to find the periphery
    eccentricity = nx.eccentricity(G)
    diameter = nx.diameter(G)
    periphery = [node for node, ecc in eccentricity.items() if ecc == diameter]
    return periphery
```

```python
def find_center(G):
    #using eccentricity to find the center
    eccentricity = nx.eccentricity(G)
    radius = nx.radius(G)
    center = [node for node, ecc in eccentricity.items() if ecc == radius]
    return center
```

```python
for graph, graph_name in zip(graphs, graph_names):
    periphery = find_periphery(graph)
    center = find_center(graph)
    print(f'{graph_name} - Center: {center}')
    print(f'{graph_name} - Periphery: {periphery}\n')
```
